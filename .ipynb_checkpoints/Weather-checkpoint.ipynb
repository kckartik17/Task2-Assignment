```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('testset.csv',parse_dates=['datetime_utc'])
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df
```

```python
df.drop([' _heatindexm',' _precipm',' _wgustm',' _windchillm'],inplace=True,axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df[' _dewptm'].fillna((df[' _dewptm'].mean()), inplace=True)
df[' _hum'].fillna((df[' _hum'].mean()), inplace=True)
df[' _pressurem'].fillna((df[' _pressurem'].mean()), inplace=True)
df[' _tempm'].fillna((df[' _tempm'].mean()), inplace=True)
df[' _vism'].fillna((df[' _vism'].mean()), inplace=True)
df[' _wdird'].fillna((df[' _wdird'].mean()), inplace=True)
df[' _wspdm'].fillna((df[' _wspdm'].mean()), inplace=True)
```

```python
df[' _wdire'].fillna((df[' _wdire'].mode()[0]), inplace=True)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df
```

```python
df[' _conds'].dropna(inplace=True)
```

```python
cols = [' _dewptm', ' _fog', ' _hail', ' _hum',
       ' _pressurem', ' _rain', ' _snow', ' _tempm', ' _thunder', ' _tornado',
       ' _vism', ' _wdird', ' _wspdm']

for x in cols:
    df[x] = df[x][df[x].between(df[x].quantile(.025),df[x].quantile(.975))]
```

```python
def extract_month(date):
    return date.month
```

```python
df['Month'] = df.apply(lambda x:extract_month(x.datetime_utc),axis=1)
```

```python
df.head(5)
```

```python
df[' _dewptm'].dropna(inplace=True)
df[' _hum'].dropna(inplace=True)
df[' _pressurem'].dropna(inplace=True)
df[' _tempm'].dropna(inplace=True)
df[' _vism'].dropna(inplace=True)
df[' _wdird'].dropna(inplace=True)
df[' _wspdm'].dropna(inplace=True)
```

```python
df.head(5)
```

```python
humi = df.groupby("Month").mean()

plt.figure(figsize=(20,20))

plt.subplot(4,4,1)
plt.plot(humi[' _dewptm'])
plt.title('_dewptm')

plt.subplot(4,4,2)
plt.plot(humi[' _fog'])
plt.title('_fog')

plt.subplot(4,4,3)
plt.plot(humi[' _hail'])
plt.title('_hail')

plt.subplot(4,4,4)
plt.plot(humi[' _hum'])
plt.title('_hum')

plt.subplot(4,4,5)
plt.plot(humi[' _pressurem'])
plt.title('_pressurem')

plt.subplot(4,4,6)
plt.plot(humi[' _rain'])
plt.title('_rain')

plt.subplot(4,4,7)
plt.plot(humi[' _snow'])
plt.title('_snow')

plt.subplot(4,4,8)
plt.plot(humi[' _tempm'])
plt.title('_tempm')

plt.subplot(4,4,9)
plt.plot(humi[' _thunder'])
plt.title('_thunder')

plt.subplot(4,4,10)
plt.plot(humi[' _tornado'])
plt.title('_tornado')

plt.subplot(4,4,11)
plt.plot(humi[' _vism'])
plt.title('_vism')

plt.subplot(4,4,12)
plt.plot(humi[' _wdird'])
plt.title('_wdird')

plt.subplot(4,4,13)
plt.plot(humi[' _wspdm'])
plt.title('_wspdm')
```
